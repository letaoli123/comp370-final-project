```python
import pandas as pd

df1 = pd.read_json("../data/first_200_annotated.json")
df2 = pd.read_json("../data/manual_300_annotated.json")

merged_df = pd.concat([df1, df2], ignore_index=True)
merged_df.to_json('../data/full_annotated.json', orient='records', indent=4)
```

```python
import pandas as pd
import json
from collections import defaultdict

with open("../data/clean_annotated_articles.json", "r") as f:
    data = json.load(f)

articles_by_topic = defaultdict(list)
for article in data:
    articles_by_topic[article["Topic"]].append(article)

articles_by_topic = dict(articles_by_topic)

with open ("../data/articles_by_topic.json", "w") as f:
    json.dump(articles_by_topic, f, indent=4)

```

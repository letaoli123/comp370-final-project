```python
import pandas as pd
import json
from pathlib import Path
import os
from collections import defaultdict

root_dir = Path(os.getcwd()).parent.parent
annotated_data_path = "data/data_annotation"
annotated_file = "clean_annotated_articles.json"

annotated_file_path = root_dir / annotated_data_path / annotated_file
output_file_path = root_dir / "data" / "data_annotation" /"articles_by_topic.json"

with open(annotated_file_path, "r") as f:
    data = json.load(f)

articles_by_topic = defaultdict(list)
for article in data:
    articles_by_topic[article["Topic"]].append(article)

articles_by_topic = dict(articles_by_topic)

with open (output_file_path, "w") as f:
    json.dump(articles_by_topic, f, indent=4)

```
